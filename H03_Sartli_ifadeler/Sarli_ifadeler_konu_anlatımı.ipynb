#### IF Ifadeleri

En basit şartlı idafeleri if keyword ile oluşturuyoruz

Bir havanın sıcaklık durumuna göre bilgi yazdıran bir fonksiyon içerisinde bunu inceleyeyim.

```python
def havayi_degerlendir(hava_sicakligi:float):
    """
    Sıcaklığın derecesine göre mesaj döndüren fonksiyon
    
    Args:
        hava_sıcakligi (float): Anlik hava sicakligi 

    Returns:
        message (str): Hava durmunun metinsel karşılığı
    """

    message = "Normal"
    if hava_sicakligi > 38.0:
        message = "Sıcak! Çok sıcak! Dahada sıcak olacak!"
    return message

print(havayi_degerlendir(38.5))
print(havayi_degerlendir(38.0))
print(havayi_degerlendir(38))
```

##### IF ... else ifadeleri

Eğer istenirse if'den sonra tek bir ihtimal kalmasıdurumunda (bu zorunlu) else keywordu kullanılabilir. Else'den sonra aynı durumu bağlayan başka herhangi bir şart oluşturulamaz.

```python
def eval_temp(temp:float):
    """
    Compares and returns a value based on weather temperature given as `temp` parameter.

    Args:
        temp (float): Current temperature in float type.mro
    
    Returns:
        message (str): String representation of the current status.
    """
    assert isinstance(temp,float),"Parameter `temp` must be type float."
    if temp >= 38.0:
        message = "High temperature!"
    else:
        message = "Normal temperature."
    
    return message

print(eval_temp(35.0))
```

##### IF...ELIF...ELSE

Eğer ikiden fazla şartlı ifademiz varsa bunu `elif` kullanarak arttırabiliriz. Bunun faydası daha complex işler yapabilmektir.

```python
# if temp > 38.0 --> yüksek sıcaklık döndürsün

# elif temp > 36.0 --> yüksek sıcaklık döndürsün

# if temp  38.0 --> yüksek sıcaklık döndürsün

def eval_temp(temp:float):
    """
    Compares and returns a value based on weather temperature given as `temp` parameter.

    Args:
        temp (float): Current temperature in float type.mro
    
    Returns:
        message (str): String representation of the current status.
    """
    assert isinstance(temp,float),"Parameter `temp` must be type float."
    temp = round(temp,2)
    if temp > 38.0:
        message = "High temperature!"
    elif temp > 36.0:
        message = "Normal temperature!"
    else:
        message = "Low temperature."
    
    message = f"Temperature was {temp}. " + message
    return message

import random
kek = [print(eval_temp(float(random.random()*50))) for i in range(10)]
```

```python
import random
def calc_tax(wage:float) -> float:
    """
    Calculates the tax based on wage.

    Args:
        wage (float): The wage of a person to calculate tax.

    Returns:
        tax (float): Returns a float value of the tax to be paid.

    """
    assert isinstance(wage,float),"Parameter wage has to be type float."
    ratio = 0.25
    if wage >= 12000.0:
        ratio = 0.3

    tax:float = wage*ratio

    return tax
    # return f"Person's wage was: ${wage}\nTherefore tax ratio is: {ratio}\nAnd the tax is calculated as: ${round(tax,2)} *rounded to have 2 decimals*"

lol = [print(calc_tax(float(random.randrange(10000,15000))),end="\n\n") for i in range(5)]
```
